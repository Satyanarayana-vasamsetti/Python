```python
protein_content = {
    "Egg": 13,
    "Chicken Breast": 31,
    "Fish": 22,
    "Milk": 3.4,
    "Curd (Yogurt)": 10,
    "Paneer": 18,
    "Cheese": 25,
    "Almonds": 21,
    "Peanuts": 26,
    "Soybeans": 36,
    "Lentils (Dal)": 9,
    "Chickpeas": 19,
    "Kidney Beans": 24,
    "Oats": 17,
    "Brown Rice": 7.5,
    "Wheat": 13,
    "Tofu": 8,
    "Pumpkin Seeds": 30,
    "Sunflower Seeds": 21
}
s2 = pd.Series(protein_content,name = "Protein")
s2
```

```python
# Conditional Selection
s2>6
```

```python
s2[s2>20]
```

```python
# Logical Operators
s2[(s2>25) & (s2<=30)]
```

```python
s2[~(s2<20)]
```

```python
# Modifying the series
s2["Egg"] = 10
s2
```
