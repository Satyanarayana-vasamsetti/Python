## **Array Reshaping**

```python
# reshape -> it can change dimensions of a array
arr = np.array([1,2,3,4,5,6])
new_arr = arr.reshape(2,3)
new_arr
```

```python
new_arr = arr.reshape(3,2)
new_arr
```

```python
# ravel() is used to convert array into 1D array
r = new_arr.ravel()
r
r[0]=100
r
new_arr
# ravel() can change original array when we try to change ravel() array
```

```python
# flatten it is also used to covert array into 1D arrya, but it make copy of arrya
# if you can change value of flatten it can't effect on original arrya
flat = new_arr.flatten()
flat
flat[0]=1
flat
# new_arr
```

## **Arthimetic Operations on Arrays**

```python
arr = np.array([1,2,3,4,5])
ind = [2,4]
print(np.take(arr,ind))

```

```python
# nditer get only values at index
for i in np.nditer(arr):
  print(i,end = " ")
```

```python
# where as ndenumerator gives you values along with indexs
for ind,val in np.ndenumerate(arr):
  print(ind,val)
```

```python
# Views vs Copies
# View can change value even  it is original
arr = np.array([1,2,3,4,5])
view  = arr.view()
view[0] = 100
print(view)
print(arr)
```

```python
# copy() can't change original arr
arr = np.array([1,2,3,4,5])
copy  = arr.copy()
copy[0] = 100
print(copy)
print(arr)
```

```python
# The below are the aggregation functions of numpy
arr = np.array([1,2,3,4,5])
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.sum(arr))
print(np.var(arr))
print(np.average(arr))
print(np.cumsum(arr))
print(np.cumprod(arr))
```
