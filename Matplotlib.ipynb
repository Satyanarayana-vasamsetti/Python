```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
data = {
    "Salary": [
        25000, 30000, 37000, 28000, 39000,
        48000, 55000, 52000, 35000, 42000,
        46000, 50000, 58000, 60000, 45000,
        47000, 53000, 62000, 68000, 75000
    ]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
# Line plot
plt.plot(df["Salary"],color = "red",marker="o",linestyle=":",linewidth="1")
plt.grid()
plt.show()
```

```python
# Histogram
plt.hist(df['Salary'],bins = 5)
plt.show()
```

```python
df['dept'] = [
        "HR", "IT", "Finance", "HR", "IT",
        "Finance", "IT", "Marketing", "HR", "Finance",
        "IT", "Marketing", "Finance", "IT", "HR",
        "Marketing", "Finance", "IT", "Marketing", "IT"
    ]
```

```python
df['dept'].value_counts()
```

```python
plt.pie(df['dept'].value_counts(),labels = df['dept'].value_counts().index,autopct="%1.1f",explode=[0,0.1,0,0])
plt.show()
```

```python
# count plot
plt.bar(df['dept'].value_counts().index,df['dept'].value_counts())
plt.show()
```
