```python
# add new column avg_score by (G1+G2+G3)/3
df["avg_score"] = (df['G1']+df['G2']+df['G3'])/3
df
```

```python
# performance_level
# High→avg_score≥15,Medium→10–14 and Low→< 10
df["performance_level"] = np.where(df["avg_score"] >= 15, "High",np.where(df["avg_score"] >= 10, "Medium", "Low"))
df
```

```python
df["performance_level"].value_counts()
```

```python
low_alchol = df[(df["Dalc"] <= 2) & (df["Walc"] <= 2)]
high_alchol = df[(df["Dalc"] >= 3) | (df["Walc"] >= 3)]
print(low_alchol)
print(high_alchol)
```
